# Hybrid Course Recommendation

This notebook combines content-based and popularity-based recommendations.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Load dataset
courses = pd.read_csv("../data/udemy_course_data.csv")

# Content-based setup
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(courses["course_title"])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def recommend_course(idx, top_n=5):
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    return [courses.iloc[i[0]]["course_title"] for i in sim_scores[1:top_n+1]]

def popular_courses(top_n=5):
    return courses.sort_values("num_subscribers", ascending=False).head(top_n)["course_title"].tolist()

def hybrid_recommend(idx, top_n=5):
    content_recs = recommend_course(idx, top_n*2)
    pop_recs = popular_courses(top_n*2)
    final_recs = list(set(content_recs + pop_recs))[:top_n]
    return final_recs

hybrid_recommend(10)
```
